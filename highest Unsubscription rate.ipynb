```python
rate=pd.DataFrame({'Active': active, 'Unsubscribed': unsub})
rate=rate.reset_index()
```

```python
rate['list']= rate['index']
```

```python
rate.drop('index',axis=1,inplace=True)
```

```python
rate['Rate']= (rate['Unsubscribed']/(rate['Active']+rate['Unsubscribed']))*100

rate.sort(['Rate'], ascending=[False])
```

```python
rate.plot(x='list',y='Rate',kind='bar',color='purple',figsize=(15,10))
```

Nautical Novels has highest Unsubscription rate followed by Gardening and Greenery
